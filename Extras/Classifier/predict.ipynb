```python
#Prediction
def predict(model, files, dest_list):
    total = len(files)
    print('Total file to Predict :',total)
    
    results = dict()
    
    for index, file in enumerate(files):
        try:
            print(index+1,'/',total,end = ' -> ')
            result = model.single_predict(file)
            type = result[0][0]
            results[file] = type
            filename = file.split('/')[-1]
            if type < 0.5:
                print('FILE :',filename,' -> PREDICTION :','other')
                os.rename(file, dest_list[0] + '/' + filename)
            else:
                print('FILE :',filename,' -> PPREDICTION :','pathology')
                os.rename(file, dest_list[1] + '/' + filename)
        except:
            print('exception',file)
        
    return results
```

```python
weights_file = 'weights/set2_2.h5'
model = Model(weights_file)
# model.load_model(weights_file)
```

```python
files = bf.get_files('../../Dataset/mixed')
print(len(files))
```

```python
results = predict(model, files[0:len(files)], ['../../Dataset/temp_other', '../../Dataset/temp_pathology'])
```
